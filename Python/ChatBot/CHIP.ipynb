```python
def reflect(text: str) -> str: # Replace phrases using REFLECTIONS, longest-first, with word boundaries where possible.

    s = text.lower()

    # Sort keys longest->shortest so "my code" is replaced before "my"
    keys = sorted(REFLECTIONS.keys(), key=len, reverse=True)

    for k in keys: # for each key in keys
        v = REFLECTIONS[k] # value is the reflection of Key

        # If key contains letters/numbers, use word boundaries; otherwise replace
        if re.search(r"[A-Za-z0-9']", k):
            pattern = r"\b" + re.escape(k) + r"\b"
            s = re.sub(pattern, v, s)
        else:
            s = s.replace(k, v)

    return s
```

```python
EXIT_WORDS = {"exit", "quit", "bye", "goodbye"}

def chip_chat():
    #INTRODUCTION
    print("CHIP: Hello! You can call me Sr. Professor Dev CHIP!")
    print("CHIP: I’m here for a practice technical interview.")
    input("CHIP: Press Enter when you're ready...\nYou: ")

    name = input("CHIP: Let's start with your first name.\nYou: ").strip() or "there"

    print(f"CHIP: On a scale of 1–10, how would you rate your Git or Socket Programming skills, {name}?")
    rating_raw = input("You: ")
    match = re.search(r"\b([1-9]|10)\b", rating_raw)
    comfort = int(match.group()) if match else 5

    # Comfort Logic
    if comfort >= 8:
        opener = "Excellent."
    elif comfort >= 5:
        opener = "Great."
    elif comfort >= 3:
        opener = "Not bad."
    else:
        opener = "No worries."

    # language selection
    language = None
    while language is None:
        lang_raw = input(f"CHIP: {opener} Choose a language to focus on (Java / Python).\nYou: ").lower()
        m = re.search(r"\b(java|python)\b", lang_raw)
        if m: # if found language
            language = m.group(1) # set "language" to that value
        else:
            print("CHIP: I only support Java or Python right now.\n")

    print(f"CHIP: Awesome. We'll do {language.capitalize()}.")
    print(f"CHIP: Now, {name}—tell me briefly about your previous software development experience.")

    current_options = {**TECHNICAL_SCRIPTS[language], **TECHNICAL_SCRIPTS["general"]} # Access each index into TECHNICAL_SCRIPTS (Uses dictionary unpacking (**) to merge them into one dictionary)

    while True:
        user_input_raw = input("You: ").strip()

        if not user_input_raw:
            print("CHIP: Give me a sentence or two—anything to start with.")
            continue

        if any(w in user_input_raw.lower() for w in EXIT_WORDS):
            print("CHIP: Thank you for your time. My assistant will see you out now.")
            break

        # Build one response per user message
        response = None

        #  Try pattern matches first
        for pattern, responses in current_options.items():
            if re.search(pattern, user_input_raw, re.IGNORECASE):
                response = random.choice(responses)
                break

        #  If nothing matched, it chooses fallback (Eliza reflection or random technical)
        # If no pattern matched: Reflects the user's input using the reflect() function
        # 50% chance: Uses the reflection in a therapeutic-style question
        # 50% chance: Picks a random technical question with a transition phrase
        if response is None:
            reflected = reflect(user_input_raw)

            # Pick one fallback style
            if random.random() < 0.50:
                response = f'When you say "{reflected}", what part mattered most to you?' #Reflection
            else:
                pool = []
                for q_list in TECHNICAL_SCRIPTS[language].values():
                    pool.extend(q_list)

                transition = random.choice(TRANSITIONS)
                question = random.choice(pool)
                response = f"{transition} {question}" # Technical Response

        print(f"CHIP: {response}")

if __name__ == "__main__":
    chip_chat()
```
